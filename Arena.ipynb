```python
class Arena:
    def __init__(self, Bot1, Bot2):
        self.bbot1 = Bot1
        self.bbot2 = Bot2
    def battle(self):
        while self.bbot1.is_alive() and self.bbot2.is_alive():
            #begin battle round
            if self.bbot1.speed >= self.bbot2.speed:
                self.bbot1.action(self.bbot2)
                self.bbot2.action(self.bbot1)
            else:
                self.bbot2.action(self.bbot1)
                self.bbot1.action(self.bbot2)
            self.bbot1.get_stats()
            self.bbot2.get_stats()
            if input("Type 'n' to end game") == 'n':
                return
        if self.bbot1.is_alive():
            print(self.bbot1.name + " is the winner!")
        else:
            print(self.bbot2.name + " is the winner!")
```

```python
import random

'''
damage_dealt
= robotBaseDamage - (robotBaseDamage * opponentBaseArmor / 100)
= robotBaseDamage * (1 - opponentBaseArmor / 100)

opponentBaseArmor / 100 is like the “armor absorption fraction”

Example: If you get 40 damage, but you have 75 armor, the absorption fraction is 75/100 = 3/4.
So the damage dealt should be 40 * (1 - 3/4) = 40 * 1/4 = 10
'''

class Jan:
	def __init__(self, name):
		self.name = name
		self.health = 100.0
		self.base_armor = 10.0
		self.base_damage = 10.0
		self.speed = 10.0

	def attack(self, opponent):
		damage_dealt = self.base_damage * (1 - opponent.base_armor / 100)
		opponent.take_damage(damage_dealt)

	def take_damage(self, damage_dealt):
		self.health -= damage_dealt

	def build_attack(self):
		
		self.base_armor -= 1
		self.base_damage += 2
		self.speed -= 1

	def build_armor(self):
		self.base_armor += 2
		self.base_damage -= 1
		self.speed -= 1

	def build_speed(self):
		self.base_armor -= 1
		self.base_damage -= 1
		self.speed += 2

	def is_alive(self):
		if self.health <= 0:
			return False
		else:
			return True

	def get_stats(self):
		print(self.name)
		print("Attack:", self.base_damage)
		print("Armor:", self.base_armor)
		print("Speed:", self.speed)
		print("Health:", self.health)

	def action(self, opponent):
		random_num = random.randint(0, 100)
		if random_num <= 20: 
			self.build_armor()
		elif random_num <= 4: 
			self.build_attack()
		elif random_num <= 4: 
			self.build_speed()
		elif random_num <= 5: 
			self.attack(opponent)
		else: 
			print(self.name + " Over!")
```

```python
class Jason:
	def __init__(self, name):
		self.name = name
		self.health = 100.0
		self.base_armor = 10.0
		self.base_damage = 10.0
		self.speed = 10.0

	def attack(self, opponent):
		damage_dealt = self.base_damage * (1 - opponent.base_armor / 100)
		opponent.take_damage(damage_dealt)

	def take_damage(self, damage_dealt):
		self.health -= damage_dealt

	def build_attack(self):
		# shorthand for self.base_armor = self.base_armor - 1
		self.base_armor -= 1
		# shorthand for self.base_damage = self.base_damage + 2
		self.base_damage += 2
		self.speed -= 1

	def build_armor(self):
		self.base_armor += 2
		self.base_damage -= 1
		self.speed -= 1

	def build_speed(self):
		self.base_armor -= 1
		self.base_damage -= 1
		self.speed += 2

	def is_alive(self):
		# This is a shorter way to do the same thing v
		# return self.health > 0
		if self.health <= 0:
			return False
		else:
			return True

	def get_stats(self):
		print(self.name)
		print("Attack:", self.base_damage)
		print("Armor:", self.base_armor)
		print("Speed:", self.speed)
		print("Health:", self.health)

	def action(self, opponent):
		random_num = random.randint(0, 100)
		if random_num <= 25: # 20% for build armor
			self.build_armor()
		elif random_num <= 45: # 20% for build attack
			self.build_attack()
		elif random_num <= 60: # 20% for build speed
			self.build_speed()
		elif random_num <= 90: # 30% for attack
			self.attack(opponent)
		else: 
			print(self.name + " glitched out!")
```
